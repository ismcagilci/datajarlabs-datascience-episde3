## numpy kütüphanesini kullanarak üreteceğiniz 3 farklı veri aracılığı ile merkezi eğilim ve dağılım ölçümlerini hem yukarıda verilen matematiksel formülle hem de python'da hazır bulunan (built-in) kodlar aracılığıyla analiz edin!!!

```python
df = pd.DataFrame()
```

```python
df["Notlar"] = np.append(np.random.normal(75,10,10),np.random.normal(65,10,10))
```

```python
df
```

```python
sınıf_notları = np.append(np.random.normal(75,10,10),np.random.normal(65,10,10))
```

```python
sınıf_notları
```

```python
## Normal hesaplama
sınıf_notları_ortalama = sum(sınıf_notları)/len(sınıf_notları)
```

```python
sınıf_notları_ortalama
```

```python
##Hazır fonksiyon temelli hesaplama
df["Ortalama"] = sınıf_notları.mean()
```

```python
df
```

```python
df["Mode"] = df["Notlar"].mode()
```

```python
df
```

```python
df["Median"] = df["Notlar"].median()
```

```python
df
```

```python
df["Varyans"] = df["Notlar"].var()
df["Standart Sapma"] = df["Notlar"].std()
df["Standart Hata"] = np.std(sınıf_notları ,ddof=1) / np.sqrt(len(sınıf_notları))
```
