## Bu soruda, merkezi limit teoreminin çalışıp çalışmadığını test edeceksiniz. İki adet normal dağılıma sahip 1000 gözlem içeren değişken üreteceksiniz. Bu değişkenlerin, ortalama ve standart sapmasını kendiniz belirleyebilirsiniz. Yapmanız gereken, ilk değişkeni 50 kere üretin ve her seferinde bu değişkenin ortalamasını alın. İkinci değişkeni ise 1000 kere üretin ve yine her seferinde bu değişkenin ortalamasını alın. Daha sonra, iki değişkenin hesapladığınız ortalamalarının histogramını çizdirin. Hangi değişkenin ortalamalarının dağılımı normal dağılıma daha yakın? Sizce Merkezi Limit Theoremi çalışmış görünüyor mu?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
variable1_mean = []
variable2_mean = []
for i in range(50):
    variable1 = np.random.normal(10,5,1000)
    variable1_mean.append(variable1.mean())
for i in range(1000):
    variable2 = np.random.normal(20,10,1000)
    variable2_mean.append(variable2.mean())
```

```python
plt.figure(figsize=(10,2),dpi=200)
plt.subplot(1,2,1)
plt.hist(variable1_mean)

plt.subplot(1,2,2)
plt.hist(variable2_mean)

plt.show()
```

```python
print("Seçilen örneklem arttıkça merkezi limit teoreminin çalıştığını gözleyebiliyoruz")
```

```python
#Q-Q grafiğinde görelim
from scipy import stats

plt.figure(figsize=(10,2),dpi=200)


plt.subplot(1,2,1)
stats.probplot(variable1_mean, plot=plt)
plt.title("Örneklem1 Q-Q")
plt.ylabel("Örneklem1")

plt.subplot(1,2,2)
stats.probplot(variable2_mean, plot=plt)
plt.title("Örneklem2 Q-Q")
plt.ylabel("Örneklem2")


plt.show()
```

```python
print("Q-Q grafiğinde baktığımızda örneklem2 den alınan n=1000 ortalamaların örneklem1 n=50 e göre normal dağılıma daha çok yaklaştığını görüyoruz.")
```

## İki madeni paramızın olduğunu varsayalım. Bunlardan bir tanesi adil (yani, yeteri kadar deneme yapıldığında yazı veya tura 1/2 olasılığına yaklaşmakta), diğeri ise adil değil ve atıldığında 3/4 olasılığında tura elde edilmektedir. Herhangi biri seçilip, n kere deneme yapıldığında, Büyük Sayılar Yasası uzun vadede gelecek tura sayısını tahmin etmemize izin verir mi? Çok sayıda deneme yapıldığında, hangi paranın seçildiğini bilebilir miyiz?

```python
#Paraları ihtimallere göre ayarlayalım
money_binom1 = np.random.binomial(1,0.5,1000)
money_binom2 = np.random.binomial(1,0.75,1000)
```

```python
#İçlerinden n=10 örneklem seçelim
money_binom1_sample_10 = money_binom1[:10]
money_binom2_sample_10 = money_binom2[:10]
```

```python
#Bu örneklemleri histogram grafiği olarak çizdirelim
plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.hist(money_binom1_sample_10)

plt.subplot(1,2,2)
plt.hist(money_binom2_sample_10)
```

## #Histogram grafiklerine baktığımızda yakın değerler görüyoruz, hileli parayı ayırt etmek zor ve hatalı olur bu yüzden alınan sample miktarını arttırıp yeniden deniyoruz.

```python
#İçlerinden n=100 örneklem seçelim
money_binom1_sample_100 = money_binom1[:100]
money_binom2_sample_100 = money_binom2[:100]
```

```python
#Bu örneklemleri histogram grafiği olarak çizdirelim
plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.hist(money_binom1_sample_100)

plt.subplot(1,2,2)
plt.hist(money_binom2_sample_100)
```

## Alınan örnek miktarı arttıkça, her iki paranın da kendi ihtimallerine göre ayrıştığını çok daha rahat görebiliyoruz, bu tablolar üzerinden tahminde bulunmak çok daha doğru olacaktır.

## Son olarak tüm grafikleri yeniden görelim.

```python
plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.hist(money_binom1_sample_10)
plt.title("Normal para 10 örnek")

plt.subplot(1,2,2)
plt.hist(money_binom1_sample_100)
plt.title("Normal para 100 örnek")

plt.show()

plt.figure(figsize=(10,5),dpi=200)

plt.subplot(1,2,1)
plt.hist(money_binom2_sample_10)
plt.title("Hileli para 10 örnek")

plt.subplot(1,2,2)
plt.hist(money_binom2_sample_100)
plt.title("Hileli para 100 örnek")

plt.show()
```
